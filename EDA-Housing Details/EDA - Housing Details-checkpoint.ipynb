```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots
from seaborn import scatterplot
```

```python
df = pd.read_csv('D:/PY_projects/Housing_Details.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['households']= pd.to_numeric(df['households'],errors='coerce')
df['households'].info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
fig,ax=subplots(ncols=2,figsize=(13,5))
df[['housing_median_age']].plot.hist(bins=52,ax=ax[0])
scatterplot(data=df,x='latitude' ,y='longitude' , hue= 'housing_median_age',s=20, ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df[['total_rooms']].plot.hist(bins=100,ax=ax[0],log=True)
scatterplot(data=df, x='longitude', y ='latitude',hue='total_rooms',s=20,ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df[['total_bedrooms']].plot.hist(bins=100,ax=ax[0],log=True)
scatterplot(data=df, x='longitude', y ='latitude',hue='total_bedrooms',s=20,ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df[['population']].plot.hist(bins=100,ax=ax[0],log=True)
scatterplot(data=df, x='longitude', y ='latitude',hue='population',s=20,ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df[['households']].plot.hist(bins=100,ax=ax[0],log=True)
scatterplot(data=df, x='longitude', y ='latitude',hue='households',s=20,ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df[['median_income']].plot.hist(bins=100,ax=ax[0],log=True)
scatterplot(data=df, x='longitude', y ='latitude',hue='median_income',s=20,ax=ax[1])
```

```python
fig,ax=subplots(ncols=2,figsize=(12,6))
df['ocean_proximity'].value_counts().plot(ax=ax[0],kind='bar')
scatterplot(data=df, x='longitude', y ='latitude',hue='ocean_proximity',s=20,ax=ax[1])
```

```python
from seaborn import heatmap
heatmap(data=df.corr())
```

```python
fig,ax=subplots(ncols=2,figsize=(13,7))
df['gender'].fillna('unknow').value_counts().plot(kind='pie',ax=ax[0])
scatterplot(ax=ax[1],data=df,x='longitude', y ='latitude',hue='gender',s=20)
```

```python
from seaborn import pairplot
pairplot_columns = ['housing_median_age', 'total_rooms','ocean_proximity']
pairplot(data=df[pairplot_columns],hue='ocean_proximity')
```
