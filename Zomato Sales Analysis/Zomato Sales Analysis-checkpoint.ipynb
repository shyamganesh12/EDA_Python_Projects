```python
plt.figure(figsize=(12,6))
top_Locations=df["location"].value_counts()[:10]
sns.barplot(x=top_Locations.index,y=top_Locations.values,orient='v')
plt.title("Top 10 Restaurant Locations")
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=25)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_cuisines = df['cuisines'].value_counts()[:20]
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, orient='h')
plt.title('Top 20 Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisine')
plt.show()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
top_votes=df['votes'].value_counts()[:20]
sns.barplot( x=top_votes.index, y=top_votes.values)
plt.title('Top Votes by Location')
plt.xlabel('Location')
plt.ylabel('Top votes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='location', y='rating')
plt.title('Ratings Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(3, 6))
sns.barplot(data=df, x='online_order', y='rating')
plt.title('Ratings Comparison: Chain vs. Individual Restaurants')
plt.xlabel('Online_Order')
plt.ylabel('Rating')
plt.show()
```

```python
from wordcloud import WordCloud

# Word cloud for restaurant names
name_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['name']))
plt.figure(figsize=(10, 5))
plt.imshow(name_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Restaurant Names')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
rating_counts = df['rating'].value_counts()
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Count of Restaurants by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_locations = df['location'].value_counts()[:15]
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.title('Top 15 Restaurant Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
restaurants_with_online_order = df[df['online_order'] == 'Yes']
sorted_restaurants = restaurants_with_online_order.sort_values(by='rating', ascending=False)
top_restaurant = sorted_restaurants.iloc[0]
plt.figure(figsize=(3, 6))
sns.barplot(x=[top_restaurant['name']], y=[top_restaurant['rating']])
plt.title('Highest Rated Restaurant with Online Order')
plt.xlabel('Restaurant Name')
plt.ylabel('Rating')
plt.show()
```
